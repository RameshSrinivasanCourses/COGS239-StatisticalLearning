### Oscillations 

* Oscillations are ubiquitous in the world.  If we look at any signal, one of the most commonly observed features, is an oscillation. 

* Colloquially, we might say a signal is oscillation if we observe an alternation between peaks and troughs. 

* We need some math and physics to guide us here. 

* Sin and cosine functions defined over time,
 $ e = A sin(2 \pi f t)$
    *  t = vector of time points 
    *  f = frequency 
    *  A = amplitude 

```python
t = np.arange(0,1,0.01)
f = 5
A = 1 
e = A*np.sin(2*np.pi*f*t)

plt.plot(t,e)
plt.plot(t,e,'ro')
plt.xlabel('Time(sec)')
plt.grid(True)
plt.show()
```

* What are the units of f? 
* Frequency is the **rate of change of phase**. 

* Going back to the 3D plot, its how fast you go around the circle.  

 $2\pi f = \frac{\Delta\phi}{\Delta t}$

   * Frequency f in in units of Hz (1/sec).  
   * Frequency $w = 2 \pi f$ is in units radians/sec

### Digital Sampling 
![DIGITIZATION](img/ADC.png)

* In order to measure a physical signal and record it onto a computer, I have to perform to Analog to Digital Conversion (ADC) 

* For example, when you speak into a microphone on your phone or computer, the sound pressure waveform you are generating has to be digitized before transmission by an electronic device. 

* For example, to make the plots in the simulations above I had to represent the signals in digital form.  I defined t at discrete points in time.  

* Every digital signal has a sampling frequency

```python
f = 10 # this is the frequency of the signal. 
fs = 1000 # This is the frequency at which I sample it . 
dt = 1/fs # This is the sampling interval 
A = 1 
t = np.arange(0,1,dt)
plt.figure(figsize = (10,10))
e = A*np.sin(2*np.pi*f*t)
plt.plot(t,e,label = 'dt = ' + str(dt))
plt.grid()
# Now lets sample it at a lower frequency 
fs = 50 
dt = 1/fs 
t2 = np.arange(0,1,dt)
e2 = A*np.sin(2*np.pi*f*t2)
plt.plot(t2,e2,'ro-',label = 'dt = ' + str(dt))
plt.legend()
plt.show()
```

### Nyquist Theorem. 

* Any signal that contains the highest frequency f must be sampled at the frequency $2\times f$

### Engineer's Nyquist 

* Any signal that contains the highest frequency f must be sampels at the frequency $2.5\times f$ 

### Ramesh's Nyquist 

* Best to keep it at $5\times f$
